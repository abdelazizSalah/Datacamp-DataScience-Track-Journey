Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

## 1. Lets get the best NYC schools with math result

```python
# set the condition
high_math_results = schools.average_math > (0.8 * 800)

# select the best schools
best_math_schools = schools[high_math_results]

# subsetting two columns only 
best_math_schools = best_math_schools[["school_name", "average_math"]]

# sort them on the average_math
best_math_schools = best_math_schools.sort_values("average_math", ascending = False)

best_math_schools.head()
```

## 2. Getting the top 10 performing schools based on the combined SAT scores
* SAT Score will be the summation of the three scores to be out of 2400 points.

```python
# evaluate the total SAT for all schools
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

# sort them in descending order 
schools = schools.sort_values('total_SAT', ascending = False)

# select the top 10 rows 
top_10_schools = schools[:10]

# selecting school name an total_SAT columns only 
top_10_schools = top_10_schools[["school_name", "total_SAT"]]

# print the total Sat schools
top_10_schools
```

## 3. Evaluating the borough with the largest std dev in the SAT score 

```python
# lets group the schools by the borough
grouped_schools = schools.groupby("borough")

# lets compute the std of total_SAT
grouped_schools_with_std = grouped_schools["total_SAT"].std()

# getting the highest borough 
highest_borough = grouped_schools_with_std.idxmax()

# now lets get all the schools in that borough 
same_borough = schools.borough == highest_borough
schools = schools[same_borough]

# getting the required values 
numOfSchools = round(schools["total_SAT"].count(), 2)
average_SAT =  round(schools["total_SAT"].mean(), 2)
SAT_stdev =    round(schools["total_SAT"].std(), 2)

largest_std_dev = pd.DataFrame ([{
    "borough" : highest_borough, 
    "num_schools" : numOfSchools,
    "average_SAT" : average_SAT, 
    "std_SAT" : SAT_stdev
    }])

largest_std_dev

```
